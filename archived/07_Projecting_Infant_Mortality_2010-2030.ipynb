```python
df_births = pd.read_excel('~/cpt5/capstone/data/processed/infant_births.xls')
```

```python
for i in np.arange(2021,2031):
    df_births[i] = np.nan
```

```python
df_births
```

```python
year = np.arange(2010,2031)
```

```python
df_births = df_births.melt(id_vars='Region', value_vars=year)
df_births.columns= ['Region', 'year', 'live_births']
```

```python
df_births
```

```python
df_births['Region'].replace({
                 'NATIONAL CAPITAL REGION (NCR)': "NCR", 
                 'CORDILLERA ADMINISTRATIVE REGION (CAR)': 'CAR',
                 'REGION I (ILOCOS REGION)': "I - Ilocos",
                 'REGION II (CAGAYAN VALLEY)': "II - Cagayan",
                 'REGION III (CENTRAL LUZON)': "III - C. Luzon",
                 'REGION IV-A (CALABARZON)': "IVA - Calabarzon",
                 ' MIMAROPA REGION': "MIMAROPA",
                 'REGION V (BICOL REGION)': "V - Bicol",
                 'REGION VI (WESTERN VISAYAS)': "VI - W. Visayas",
                 'REGION VII (CENTRAL VISAYAS)': "VII - C. Visayas" ,
                 'REGION VIII (EASTERN VISAYAS)': "VIII - E. Visayas",
                 'REGION IX (ZAMBOANGA PENINSULA)': "IX - Zamboanga",
                 'REGION X (NORTHERN MINDANAO)': "X - N. Mindanao",
                 'REGION XI (DAVAO REGION)':"XI - Davao",
                 'REGION XII (SOCCSKSARGEN)':"XII - SOCCSKSARGEN",
                 'REGION XIII (Caraga)': "XIII - Caraga",
                 'AUTONOMOUS REGION IN MUSLIM MINDANAO (ARMM)':
                                                            "BARMM"

                        }, inplace=True)
```

```python
df_infant.columns = ['year', 'Region','Male', 'Female', 'infant_count' ]
```

```python
df_infant_rate = df_infant.merge(df_births, on=['year', 'Region'], how='right')
df_infant_rate
```

```python
df_infant_rate.isna().sum()
```

```python
df_infant_rate['infant_death_rate_per100k']= (df_infant_rate['infant_count']/df_infant_rate['live_births'])*100_000
```

```python
plt.figure(figsize=(10,5))
for i in df_infant_rate['Region'].unique():
    df = df_infant_rate[df_infant_rate['Region'] == i]
    plt.plot(df['year'], df['infant_death_rate_per100k'],'+-', label=i)
plt.legend(bbox_to_anchor=(1,1))
plt.xticks(np.arange(2010,2021))
plt.title('infant mortality rate per 100k (2010-2020)');
```

```python
df = pd.DataFrame(df_infant_rate.groupby('Region')['infant_death_rate_per100k'].mean()).reset_index()
df
```

```python
yr = np.arange(2021, 2031)
```

```python
df_dummy = pd.DataFrame()
df_dummy['year'] = yr
```

```python
for i in yr:
    for reg in df_infant_rate.Region.unique():
        value = df[df['Region']==reg]['infant_death_rate_per100k'].values[0]
        
        df_infant_rate.loc[(df_infant_rate.year==i)
                 & (df_infant_rate.Region==reg),
                     'infant_death_rate_per100k'] = value
```

```python
df_infant_rate
```

```python
labels = ["NCR", "CAR", "I - Ilocos", "II - Cagayan",
          "III - C. Luzon", "IVA - Calabarzon",
          "MIMAROPA", "V - Bicol", "VI - W. Visayas",
          "VII - C. Visayas", "VIII - E. Visayas", 
          "IX - Zamboanga",
          "X - N. Mindanao", "XI - Davao", "XII - SOCCSKSARGEN",
          "XIII - Caraga", "BARMM"]

colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple',
         'tab:brown', 'tab:pink', 'indigo', 'tab:olive', 'tab:cyan',
         'darkmagenta', 'maroon', 'lime', 'darkolivegreen', 'cadetblue',
         'darkgoldenrod', 'fuchsia']
```

```python
plt.figure(figsize=(10,5))
for i,j in enumerate(labels):
    df = df_infant_rate[df_infant_rate['Region'] == j]
    if j in ['NCR', "IVA - Calabarzon", "I - Ilocos", 'BARMM']:
   
        plt.plot(df['year'], df['infant_death_rate_per100k'],'+-', label=j,
                color=colors[i], lw=2)
    else:
        plt.plot(df['year'], df['infant_death_rate_per100k'],'+-', label=j,
                color='darkgrey')
    
plt.legend(bbox_to_anchor=(1,1))
plt.xticks(np.arange(2010,2031,2))
# plt.title('Infant Mortality Rates per Region, 2010-2030', fontsize=16)
plt.ylabel('Mortality Rate (per 100,000)', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.tight_layout()
plt.savefig('PPT_infant_mortality.png', dpi=300);
```

```python
plt.figure(figsize=(10,5))
for i,j in enumerate(labels):
    df = df_infant_rate[df_infant_rate['Region'] == j]
    plt.plot(df['year'], df['infant_death_rate_per100k'],'+-', label=j,
                color=colors[i], lw=2)
    
plt.legend(bbox_to_anchor=(1,1))
plt.xticks(np.arange(2010,2031,2))
plt.title('Infant Mortality Rates per Region, 2010-2030', fontsize=16)
plt.ylabel('Mortality Rate (per 100,000)', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.tight_layout()
plt.savefig('infant_mortality.png', dpi=300);
```
